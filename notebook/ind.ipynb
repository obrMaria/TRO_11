### Индивидуальное задание

Представим, что у нас есть нечеткое изображение с QR-кодом. Нужно восстановить код так, чтобы он был читаемым.


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('qr.jpg', 0)
```

Применим пороговую обработку, чтобы уменьшить шум на изображении и выделить четкие границы QR-кода

```python
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

добавляем шум на изображение, чтобы восстановить недостающие части QR-кода

```python
# Добавляем шум на изображение
gaussian_noise = np.random.normal(0, 1, img.shape)
img_noise = img + gaussian_noise
```

снова применяем пороговую обработку, чтобы очистить изображение от дополнительного шума и убедиться, что код читается корректно

```python
ret, thresh_noise = cv2.threshold(img_noise, 127, 255, cv2.THRESH_BINARY)
```

```python
# Отображаем исходное изображение и обработанное изображение
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(thresh,cmap = 'gray')
plt.title('Thresholded'), plt.axis('off')
plt.subplot(2,2,3),plt.imshow(img_noise,cmap = 'gray')
plt.title('Noisy'), plt.axis('off')
plt.subplot(2,2,4),plt.imshow(thresh_noise,cmap = 'gray')
plt.title('Noisy Thresholded'), plt.axis('off')

plt.show()
```
